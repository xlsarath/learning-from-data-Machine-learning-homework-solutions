## Part 2 (1) Native version of my own Neural Network w.r.t  given params 
2 input layers; 
m - hidden layers; 
1 output layers; 
tanh activation function;

```python
import numpy as np

X = np.array(([1, 2], [2, 1], [5, 2]), dtype=float)
y = np.array(([92], [86], [82]), dtype=float)

xPredicted = np.array(([1,4]), dtype=float)

# scale units
X = X/np.amax(X, axis=0) # maximum of X array
xPredicted = xPredicted/np.amax(xPredicted, axis=0) # maximum of xPredicted (our input data for the prediction)
y = y/100 # max test score is 100

class Neural_Network(object):
  def __init__(self):
    #parameters
    self.inputSize = 2
    self.outputSize = 1
    self.hiddenSize = 2

    #weights
    self.W1 = np.random.randn(self.inputSize, self.hiddenSize) # (2x2) weight matrix from input to hidden layer
    self.W2 = np.random.randn(self.hiddenSize, self.outputSize) # (2x1) weight matrix from hidden to output layer

  def forward_prop(self, X):
    #forward propagation through our network
    self.z = np.dot(X, self.W1) # dot product of input x and weights
    self.z2 = self.tanh(self.z) # activation function
    self.z3 = np.dot(self.z2, self.W2) # dot product with hidden layer  and 2nd set of weights
    o = self.tanh(self.z3) # tanh activation function
    return o #return obj

  def tanh(self, s):
    # activation function
    return np.tanh(s)

  def tanhDer(self, s):
    #derivative of tanh
    return 1.0 - np.tanh(s)**2


  def backward_prop(self, X, y, o):
    # backward propgation through the network
    self.o_error = y - o # error in output
    self.o_delta = self.o_error*self.tanhDer(o) # applying derivative of tanh to error

    self.z2_error = self.o_delta.dot(self.W2.T) # z2 error: how much our hidden layer weights contributed to output error
    self.z2_delta = self.z2_error*self.tanhDer(self.z2) # applying derivative of sigmoid to z2 error

    self.W1 += X.T.dot(self.z2_delta) # adjusting first set (input --> hidden) weights
    self.W2 += self.z2.T.dot(self.o_delta) # adjusting second set (hidden --> output) weights

  def train(self, X, y):
    o = self.forward_prop(X)
    self.backward_prop(X, y, o)

  def saveWeights(self):
    np.savetxt("w1.txt", self.W1, fmt="%s")
    np.savetxt("w2.txt", self.W2, fmt="%s")

  def predict(self):
    print ("Predicted data based on trained weights: ");
    print ("Input (scaled): \n" + str(xPredicted));
    print ("Output: \n" + str(self.forward_prop(xPredicted)));

NN = Neural_Network()
for i in range(10000): # training time
  print (" #" + str(i) + "\n")
  print ("Input (scaled): \n" + str(X))
  print ("Actual Output: \n" + str(y))
  print ("Predicted Output: \n" + str(NN.forward_prop(X)))
  print ("Loss: \n" + str(np.mean(np.square(y - NN.forward_prop(X))))) # mean sum squared loss
  print ("\n")
  NN.train(X, y)

NN.saveWeights()
NN.predict()
```
